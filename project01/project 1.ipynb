```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
result=pd.read_csv('/Users/amit/Desktop/udacity/project01/results.csv')
```

```python
result.head()
```

```python
plt.plot( 't_year', 'city_avg', data=result, markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 't_year', 'global_avg', data=result, marker='', color='olive', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.legend()
```

#### In order to create a line chart with moving average we need to transform the city and global column's moving average

```python
result['moving_average_city']=result.city_avg.rolling(window=30).mean()
result['moving_average_global']=result.global_avg.rolling(window=30).mean()
```

```python
plt.plot( 't_year', 'moving_average_city', data=result, markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 't_year', 'moving_average_global', data=result, marker='', color='olive', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.legend()
```
