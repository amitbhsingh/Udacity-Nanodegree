```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
plt.scatter(data=iris,x='species',y='petal_length')
```

```python
sns.regplot(data=iris,x='sepal_length',y='petal_length',color='blue')
```

```python
plt.scatter(data=iris,x='species',y='petal_length',alpha=1/5)
```

# Heatmaps

```python
plt.hist2d(data=iris,x='petal_length',y='petal_width');
```

# Violinplot

```python
bases=sns.color_palette()[0]
sns.violinplot(data=iris,x='petal_length',y='species',color=bases)#inner=None
```

```python
sns.violinplot(data=iris,x='species',y='petal_length',color=bases)
```

# Boxplot

```python
sns.boxplot(x='petal_length',y='species',data=iris)
```

```python
sns.boxplot(x='species',y='petal_length',data=iris)
```

# Faceting

```python
g = sns.FacetGrid(data = iris, col = 'species')
g.map(plt.hist, "petal_length")
```

# Multivariate plot

```python
g = sns.FacetGrid(data = iris, hue = 'species', size = 5,palette='colorblind')
g.map(plt.scatter, 'sepal_length', 'sepal_width')
g.add_legend()
```
